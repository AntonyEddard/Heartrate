```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = 'anto_ecg.txt'
x = np.loadtxt(fname)
```

```python
print(x)
```

```python
plt.plot(x)
plt.show()
```

```python
from scipy.signal import find_peaks
```

```python
peaks, _ = find_peaks(x, height=500)
plt.plot(x)
plt.plot(peaks, x[peaks], "x")
plt.plot(np.zeros_like(x), "--", color="gray")
plt.show()
```

```python
rr_int = np.diff(peaks)
```

```python
HR = (512/rr_int)*60
```
